### Pandas란?
데이터 처리와 분석에 최적화된 라이브러리
- 행과 열(테이블 형태)로 이루어진 데이터 객체를 만들어 다룰 수 있음
- 빅데이터를 처리하고 분석하는데 편리한 함수들을 제공

#### Pandas 구조
- Series : 1차원 데이터들의 집합 (index + value)
- DataFrame : 2차원의 행과 열로 이루어진 표 형태의 데이터 집합(1차원 Series가 모여서 DF를 이룸)

### Pandas 실습 예제!!!
2015~2017년 광주광역시 범죄현황 데이터를 이용해 전년 대비 지역별 범죄 증감율을 구해보자!
- (증감률 공식 = (금년-작년)/작년*100)

#### 데이터 로드
- 인덱스 칼럼은 '관서명'으로설정할 것
- 인코딩은 'euc-kr'로 설정할 것

```python
df_2015 = pd.read_csv('data/2015.csv', encoding='euc-kr', index_col="관서명")
df_2016 = pd.read_csv('data/2016.csv', encoding='euc-kr', index_col="관서명")
df_2017 = pd.read_csv('data/2017.csv', encoding='euc-kr', index_col="관서명")
```

```python
data15 = df_2015[0::6].sum(axis=1)   #시리즈 네임이 합치면 컬럼명이 된다.
data15.name = '2015총계'
data16 = df_2016[0::6].sum(axis=1)
data16.name = '2016총계'
df_2017.drop("광주지방경찰청", inplace=True)
data17 = df_2017[0::6].sum(axis=1)
data17.name = '2017총계'
data0 = (data16 - data15)/data15*100
data0.name="2015~2016 증감률"
data00 = (data17 - data16)/data16*100
data00.name="2016~2017 증감률"
df1 = pd.concat([data15, data0, data16, data00, data17], axis=1)
df1
```

```python
#총합
df_2015["총합"] = df_2015.sum(axis=1)
df_2016["총합"] = df_2016.sum(axis=1)
df_2017["총합"] = df_2017.sum(axis=1)
```

```python
s15 = df_2015.loc[df_2015['구분']=='발생건수', '총합']
s16 = df_2016.loc[df_2016['구분']=='발생건수', '총합']
s17 = df_2017loc[df_2017['구분']=='발생건수', '총합']
```

```python
s1516 = (s16-s15)/s15*100
s1617 = (s17-s16)/s16*100
```

```python
total = pd.concat([s15, s1516, s16, s1617, s17], axis=1)
```

```python
total.columns=[ "2015총계","2015~2016증감률","2016총계","2016~2017 증감율","2017총계"]
total
```
